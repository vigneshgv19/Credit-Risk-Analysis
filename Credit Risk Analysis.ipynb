# Handling Outlier 

```python
sb.boxplot(data=data1, x="CNT_CHILDREN")
```

```python
data1=data1.drop(data1[data1["CNT_FAM_MEMBERS"] > 4.5].index, inplace= False)
```

```python
sb.boxplot(data=data1, x="CNT_FAM_MEMBERS")
```

```python
data1=data1.drop(data1[data1["DAYS_EMPLOYED"]  < -6000].index, inplace= False)
```

```python
#Quantity outlier check
Q1=data1['AMT_INCOME_TOTAL'].quantile(0.25)
Q3=data1['AMT_INCOME_TOTAL'].quantile(0.75)
IQR=Q3-Q1
```

```python
#Quantity outlier removed
data1=data1[~((data1['AMT_INCOME_TOTAL']>Q3+1.5*IQR)|(data1['AMT_INCOME_TOTAL']<Q1-1.5*IQR))]
```

```python
sb.boxplot(data=data1, x="AMT_INCOME_TOTAL")
```

```python
data1=data1.drop(data1[data1["CNT_CHILDREN"] > 2.5].index, inplace= False)
```

```python
sb.boxplot(data=data1, x="CNT_CHILDREN")
```

```python
# Converting 'DAYS_BIRTH' values from Day to Years
data1['DAYS_BIRTH'] = round(data1['DAYS_BIRTH']/-365,0)
data1.rename(columns={'DAYS_BIRTH':'AGE_YEARS'}, inplace=True)
```

```python
# Converting 'DAYS_EMPLOYED' values from Day to Years
data1['DAYS_EMPLOYED'] = abs(round(data1['DAYS_EMPLOYED']/-365,0))
data1.rename(columns={'DAYS_EMPLOYED':'YEARS_EMPLOYED'}, inplace=True)  
```

```python
# As with email_id, mobile phone, work phone wont help in credit card approval
data1 = data1.drop(["FLAG_MOBIL","FLAG_EMAIL",'FLAG_WORK_PHONE','FLAG_PHONE'],axis=1)
```

# Balancing dataset

```python
# scaling all features
from sklearn.preprocessing import MinMaxScaler
MMS = MinMaxScaler()
x_train_scaled = pd.DataFrame(MMS.fit_transform(x_train), columns=x_train.columns)
x_test_scaled = pd.DataFrame(MMS.transform(x_test), columns=x_test.columns)
```

```python
# adding samples to minority class using SMOTE
from imblearn.over_sampling import SMOTE
oversample = SMOTE()

x_train_oversam, y_train_oversam = oversample.fit_resample(x_train_scaled, y_train)
x_test_oversam, y_test_oversam = oversample.fit_resample(x_test_scaled, y_test)
```

```python
# after using SMOTE 
y_train_oversam.value_counts(normalize=True)*100
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
modelfit = model.fit(x_train_oversam,y_train_oversam)
```

```python
Predict_model0=modelfit.predict(x_test_oversam)
```

```python
confusion_matrix(y_test_oversam,Predict_model0)
```

```python
lr_acc=(accuracy_score(y_test_oversam,Predict_model0)*100,"%")
lr_acc
```
